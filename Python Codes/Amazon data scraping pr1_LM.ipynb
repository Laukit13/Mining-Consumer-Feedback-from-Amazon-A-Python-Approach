```python
# import Libraries
from selenium import webdriver
from time import sleep
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import re
import string
```

```python
#Set Path for our webdriver
path='C://msedgedriver.exe'
```

```python
#open the browser
browser = webdriver.Edge(executable_path=path)
browser.get('https://www.amazon.com/ALINK-Plastic-Bubble-Smoothie-Milkshake/product-reviews/B06XV3P2N2/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews')
```

```python
Total=[]
total=browser.find_elements("xpath","//div[@class='a-row a-spacing-base a-size-base']")
for tot in total:
    Total.append(tot.text)
print(Total)
new_items=[words for segments in Total for words in segments.split()]
Rev_Count=new_items[3:4]
print(Rev_Count)
rm = str.maketrans('', '', ',')
Rev_Count=[s.translate(rm) for s in Rev_Count]
Rev_Count=[int(x) for x in Rev_Count]
print(Rev_Count)
my_int = 10
a=int(Rev_Count[0]/my_int)
print(a)
```

```python
Ratings=[]
Names=[]
Titles=[]
Reviewz=[]
Dates=[]
Helpful=[]
Image=[]
buyer=[]
ALL=[]
for i in range(a):
    print('Sraping page',i+1) 
        
    title=browser.find_elements(By.CLASS_NAME,"review-title")[2:]
    for t in title:
        Titles.append(t.text)
        
    rating = browser.find_elements(By.CLASS_NAME, "review-rating")[2:]
    for r in rating:
        Ratings.append(r.get_attribute('textContent'))
                             
    date=browser.find_elements("xpath","//span[@class='a-size-base a-color-secondary review-date']")[2:]
    for d in date:
        Dates.append(d.text)    
                
    name=browser.find_elements("xpath","//span[@class='a-profile-name']")[2:]
    for n in name:
        Names.append(n.text)
    
    review=browser.find_elements("xpath","//span[@class='a-size-base review-text review-text-content']")
    for rev in review:
        Reviewz.append(rev.text)
        removetable = str.maketrans('', '', '\n')
        Reviews=[s.translate(removetable) for s in Reviewz]
        
#     Vpurachase=browser.find_elements("xpath","//span[@class='a-size-mini a-color-state a-text-bold']")
#     for v in Vpurachase:
#         buyer.append(v.text)
        
    for m in browser.find_elements("xpath","//div[@class='a-section celwidget']"):
        if m.find_elements(By.CLASS_NAME,"review-image-tile"):
            Image.append('Image Present')
        else:
            Image.append('Image Not Present')
    
    all=browser.find_elements("xpath","//div[@class='a-section celwidget']")
    for a in all:
        ALL.append(a.text)   
        removetable = str.maketrans('', '', '\n')
        All=[s.translate(removetable) for s in ALL]
                   
                                      
    next_Page = browser.find_element("xpath","//li[@class='a-last']")
    next_Page.click()
    sleep(3)
```

```python
for i in All:
    if "found this helpful" in i.lower():
        text =i.split("found this helpful")[0][-12:-7]
        if "one" in text.lower():
            Helpful.append("1")
        else:
            text = re.findall("\d+",text)
            Helpful.append(text[-1])
    else:
        Helpful.append("0") 
```
